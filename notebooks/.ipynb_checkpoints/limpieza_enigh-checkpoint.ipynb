```python
path_folio = 'data/ENIGH/vivienda/conjunto_de_datos/conjunto_de_datos_viviendas_enigh_2018_ns.csv'

df_folio = pd.read_csv(path_folio, usecols = ['folioviv','ubica_geo'])
df_folio = df_folio.drop_duplicates('folioviv')
print(df_folio.shape)
df_folio.head(4)
```

```python
path_ubi = 'data/ENIGH/concentrado_hogar/catalogos/ubic_geo.csv'

df_ubi = pd.read_csv(path_ubi, encoding = 'Latin-1')
print(df_ubi.shape)
df_ubi.head(4)
```

```python
path_hogares = 'data/ENIGH/hogares/conjunto_de_datos/conjunto_de_datos_hogares_enigh_2018_ns.csv'

df_hog = pd.read_csv(path_hogares, encoding = 'Latin-1')
df_hog = df_hog[df_hog['foliohog']==1]
print(df_hog.shape)
df_hog.head()
```

```python
path_viv = 'data/ENIGH/vivienda/conjunto_de_datos/conjunto_de_datos_viviendas_enigh_2018_ns.csv'

df_viv = pd.read_csv(path_viv)
df_viv.head(4)
```

```python
path_ing = 'data/ENIGH/concentrado_hogar/conjunto_de_datos/conjunto_de_datos_concentradohogar_enigh_2018_ns.csv'

df_ing = pd.read_csv(path_ing, usecols = ['folioviv','ing_cor'])
df_ing.head()
```

```python
df_location = pd.merge(df_folio, df_ubi, left_on = 'ubica_geo',
                       right_on = 'ubic_geo', how = 'inner').drop('ubic_geo', 1)

df_location.head()
```

```python
df_main = pd.merge(df_location, df_viv)
df_main = pd.merge(df_main, df_ing)
df_main = df_main[df_main['entidad']==19]
df_main.head()
```

```python
df_inegi = pd.merge(df_main, df_hog)
df_inegi.head()
```

```python
cols = ['desc_ent', 'des_mun', 'cuart_dorm', 'tv_paga', 'conex_inte', 'num_compu', 'num_lavad', 'num_refri',
       'telefono', 'num_auto','num_van','num_juego', 'ing_cor']

df_ind = df_inegi[cols]
df_ind['num_auto_correcto'] = df_ind['num_auto'] + df_ind['num_van']
df_ind.drop(['num_auto','num_van'], 1, inplace = True)
df_ind.head()
```

```python
df_ind.describe()
```

```python
df_ind['cuart_dorm'] = ['1dorm' if x==1 else '2masdorm' for x in df_ind['cuart_dorm']]

df_ind['tv_paga'] = ['TVPaga' if x==1 else 'NoTVPaga' for x in df_ind['tv_paga']]

df_ind['num_compu'] = ['PC' if x!=0 else 'NoPC' for x in df_ind['num_compu']]

df_ind['num_lavad'] = ['Lavadora' if x!=0 else 'NoLavadora' for x in df_ind['num_lavad']]

df_ind['num_refri'] = ['Refri' if x!=0 else 'NoRefri' for x in df_ind['num_refri']]

df_ind['conex_inte'] = ['Internet' if x==1 else 'NoInternet' for x in df_ind['conex_inte']]

df_ind['telefono'] = ['Telefono' if x==1 else 'NoTelefono' for x in df_ind['telefono']]

df_ind['num_auto_correcto'] = ['Auto' if x>0 else 'NoAuto' for x in df_ind['num_auto_correcto']]

df_ind['num_juego'] = ['Consola' if x>0 else 'NoConsola' for x in df_ind['num_juego']]

df_ind
```

```python
df_ind.columns = ['Municipio', 'Dorm', 'TV', 'INT',
       'PCs', 'Lavadoras', 'Refrigeradoes', 'Tel', 'Consolas',
       'Ingreso corriente', 'Autos']

df_ind = df_ind[['Municipio', 'Dorm', 'TV', 'INT',
       'PCs', 'Lavadoras', 'Refrigeradoes', 'Tel', 'Consolas',
       'Autos','Ingreso corriente']]

df_ind.head()
```

```python
df_ind.to_csv('data/Clean data/ind.csv')
```
