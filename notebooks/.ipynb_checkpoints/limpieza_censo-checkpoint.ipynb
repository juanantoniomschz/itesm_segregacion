```python
import pandas as pd
import numpy as np
import os

pd.options.display.max_columns = None
```

```python
os.chdir('/Users/juanantoniomschz/Documents/Proyectos/itesm_segregacion/')
```

```python
path_censo = '~/Documents/Proyectos/itesm_segregacion/data/Censo/RESAGEBURB_19CSV20.csv'
# path_censo = '~/Documents/Proyectos/itesm_segregacion/data/Censo/agebs_2021.csv'

df_censo = pd.read_csv(path_censo, encoding='utf-8', na_values = ['*'])
df_censo = df_censo[df_censo['NOM_LOC']=='Total AGEB urbana'].reset_index().drop('index', axis = 1)

# df_censo = pd.read_csv(path_censo, encoding='utf-8', na_values = ['*'])
df_censo.head()
```

```python
cols = ['NOM_ENT', 'NOM_MUN', 'AGEB', 'VIVPARH_CV', 'OCUPVIVPAR', 'VPH_1DOR', 'VPH_2YMASD',
        'VPH_LAVAD', 'VPH_REFRI', 'VPH_STVP', 'VPH_TELEF', 'VPH_INTER', 'VPH_PC', 'VPH_CVJ','VPH_AUTOM']

df_censo_var = df_censo[cols] # seleccionamos las columnas objetivo
df_censo_var.head(5)
```

```python
df_censo_var.columns = ['Entidad', 'Municipio', 'AGEB', 'VIVPARH_CV', 'Ocupantes', '1dorm', '2masdorm',
                        'Lavadora', 'Refri', 'TVPaga', 'Telefono', 'Internet', 'PC', 'Consola', 'Auto'] 

for i in df_censo_var.columns[3:]:
    
    df_censo_var[i] = pd.to_numeric(df_censo_var[i]) # cambiamos el formato de las columnas a numeros
    
df_censo_var.describe()
```

```python
df_cons = df_censo_var[df_censo_var['1dorm']>0] # que existan viviendas con más de 1 dormitorio en la AGEB
df_cons = df_cons[df_cons.isnull().sum(1)==0] # que la AGEB no presente valores nulos

viv_drop = (df_censo_var[df_censo_var.isnull().sum(1)>0]['VIVPARH_CV'].sum()) # cuantas viviendas dropeamos
viv_rest = df_cons['VIVPARH_CV'].sum() # cuantas viviendas quedan

print('Se eliminaron {} viviendas, quedan restantes {} viviendas.\n'.format(int(viv_drop), int(viv_rest)))

elim = len(df_censo_var) - len(df_cons)  # cuantas agebs quitamos
ageb_rest = len(df_cons) # cuantas agebs quedan

print('Se eliminaron {} agebs. Quedan restantes {} agebs'.format(elim, ageb_rest))

df_cons.describe()
```

```python
loop_var = ['Lavadora', 'Refri', 'TVPaga', 'Telefono', 'Internet', 'PC', 'Consola', 'Auto']

for i in loop_var:
    
    var = 'No' + i
    df_cons[var] = (df_cons['VIVPARH_CV'] - df_cons[i]).astype(float) # creamos variables negativas
    
df_cons.describe()
```

```python
# Cuantas AGEBS fueron registradas incorrectamente en los dormitorios
dorms_mal = len(df_cons[(df_cons.loc[:, 'VIVPARH_CV'] - df_cons.loc[:, '1dorm'] - df_cons.loc[:, '2masdorm']) != 0])
print('{} AGEBS tienen mal registrada la cantidad de dormitorios.'.format(dorms_mal))
```

```python
# Función para resolver el registro de dormitorios

def diff_num(x):
    
    '''
    x: es un dataframe con las columnas deseadas
    '''
    
    x = list(x)
    total = x.pop(0)
    
    suma = sum(x)

    for i in range(len(x)):
        x[i] = float(round((x[i]/suma) * total))
        
    return pd.Series(x)
```

```python
df_cons[['1dorm', '2masdorm']] = df_cons[['VIVPARH_CV', '1dorm', '2masdorm']].apply(diff_num, axis = 1)
df_cons[['VIVPARH_CV','1dorm', '2masdorm']].describe()
```

```python
# Cuantas AGEBS fueron registradas incorrectamente en los dormitorios
dorms_mal = len(df_cons[(df_cons.loc[:, 'VIVPARH_CV'] - df_cons.loc[:, '1dorm'] - df_cons.loc[:, '2masdorm']) != 0])
print('{} AGEBS tienen mal registrada la cantidad de dormitorios.'.format(dorms_mal))
```

```python
df_cons = df_cons[['Entidad', 'Municipio', 'AGEB', 'VIVPARH_CV', 'Ocupantes', '1dorm',
       '2masdorm','TVPaga','NoTVPaga','Internet','NoInternet', 'PC','NoPC', 'Lavadora', 'NoLavadora',
        'Refri','NoRefri','Telefono','NoTelefono','Auto','NoAuto','Consola','NoConsola']]
df_cons.reset_index(inplace = True)
df_cons.drop(['index','Entidad'], axis = 1, inplace = True)
df_cons.head()
```
