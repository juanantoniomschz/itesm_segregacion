```python
path_censo = '~/Documents/Proyectos/itesm_segregacion/data/Censo/RESAGEBURB_19CSV20.csv'

df_censo = pd.read_csv(path_censo, encoding='utf-8', na_values = ['*'])
df_censo = df_censo[df_censo['NOM_LOC']=='Total AGEB urbana'].reset_index().drop('index', axis = 1)
df_censo.head()
```

```python
df_censo['ENTIDAD'] = df_censo['ENTIDAD'].astype(str)
df_censo['MUN'] = [str(i).zfill(3) for i in df_censo['MUN']]
df_censo['LOC'] = [str(i).zfill(4) for i in df_censo['LOC']]
df_censo['CVEGEO'] = df_censo['ENTIDAD'] + df_censo['MUN'] + df_censo['LOC'] + df_censo['AGEB']
df_censo.head()
```

```python
cols = ['CVEGEO','NOM_ENT', 'NOM_MUN', 'AGEB', 'VIVPARH_CV', 'OCUPVIVPAR', 'VPH_1DOR', 'VPH_2YMASD',
        'VPH_1CUART', 'VPH_2CUART', 'VPH_3YMASC', 'VPH_LAVAD', 'VPH_REFRI', 'VPH_STVP', 'VPH_TELEF', 'VPH_INTER', 'VPH_PC', 'VPH_CVJ','VPH_AUTOM']

df_censo_var = df_censo[cols] # seleccionamos las columnas objetivo
df_censo_var.head(5)
```

```python
# Función para resolver el registro de dormitorios

def diff_num(x):
    
    '''
    x: es un dataframe con las columnas deseadas
    '''
    
    x = list(x)
    total = x.pop(0)
    
    suma = sum(x)

    for i in range(len(x)):
        x[i] = float(round((x[i]/suma) * total))
        
    return pd.Series(x)
```

```python
int_trs <- function(x){
  # For generalisation purpose, x becomes a vector
  xv <- as.vector(x) # allows trs to work on matrices
  xint <- floor(xv) # integer part of the weight
  r <- xv - xint # decimal part of the weight
  def <- round(sum(r)) # the deficit population
  # the weights be 'topped up' (+ 1 applied)
  topup <- sample(length(x), size = def, prob = r)
  xint[topup] <- xint[topup] + 1
  dim(xint) <- dim(x)
  dimnames(xint) <- dimnames(x)
  xint
}
```

```python
def diff_num_cuart(x):
    
    '''
    x: es un dataframe con las columnas deseadas
    '''
    
    x_vector = list(x)
    viviendas_tot = x_vector.pop(0)
    x_vector = np.array(x_vector)
    suma_var = sum(x_vector)
    
    x_vector = x_vector / suma_var * viviendas_tot

    xint = np.array([math.modf(i)[1] for i in x_vector])
    r = x_vector - xint
    #r = r/r.sum()
    #suma_r = sum(r)
    #topup = np.random.choice(range(len(x_vector)), round(suma_r), p = r)
    #xint[topup] = xint[topup] + 1
        
    return xint
```
