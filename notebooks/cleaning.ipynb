```python
import pandas as pd
import numpy as np

pd.options.display.max_columns = None
```

```python
path_folio = '~/Documents/Proyectos/itesm_segregacion/data/ENIGH/concentrado_hogar/conjunto_de_datos/conjunto_de_datos_concentradohogar_enigh_2018_ns.csv'

df_folio = pd.read_csv(path_folio, usecols = ['folioviv','ubica_geo'])
df_folio = df_folio.drop_duplicates('folioviv')
print(df_folio.shape)
df_folio.head(4)
```

```python
path_ubi = '~/Documents/Proyectos/itesm_segregacion/data/ENIGH/concentrado_hogar/catalogos/ubic_geo.csv'

df_ubi = pd.read_csv(path_ubi, encoding = 'Latin-1')
print(df_ubi.shape)
df_ubi.head(4)
```

```python
df_location = pd.merge(df_folio, df_ubi, left_on = 'ubica_geo', right_on = 'ubic_geo', how = 'inner').drop('ubic_geo', 1)
df_location.head()
```

```python
path_pob = '~/Documents/Proyectos/itesm_segregacion/data/ENIGH/poblacion/conjunto_de_datos/conjunto_de_datos_poblacion_enigh_2018_ns.csv'

df_pob = pd.read_csv(path_pob)
df_pob.head(4)
```

```python
clean_data = pd.merge(df_location, df_pob)
clean_data.to_csv('~/Documents/Proyectos/itesm_segregacion/data/Clean data/poblacion_enigh.csv')
clean_data[clean_data['desc_ent']=='Nuevo Leon'].to_csv('~/Documents/Proyectos/itesm_segregacion/data/Clean data/poblacion_enigh_nuevoleon.csv')
```
