```python
# Función para resolver el registro de dormitorios

def diff_num(x):
    
    '''
    x: es un dataframe con las columnas deseadas
    '''
    
    x = list(x)
    total = x.pop(0)
    
    suma = sum(x)

    for i in range(len(x)):
        x[i] = float(round((x[i]/suma) * total))
        
    return pd.Series(x)
```

```python
df_cons[['1dorm', '2masdorm']] = df_cons[['VIVPARH_CV', '1dorm', '2masdorm']].apply(diff_num, axis = 1)
df_cons[['VIVPARH_CV','1dorm', '2masdorm']].describe()
```

```python
df_cons[['1cuart', '2cuart', '3mascuart']] = df_cons[['VIVPARH_CV','1cuart', '2cuart', '3mascuart']].apply(diff_num, axis = 1)
df_cons[['VIVPARH_CV','1cuart', '2cuart', '3mascuart']].describe()
```

```python
# Cuantas AGEBS fueron registradas incorrectamente en los cuartos
cuart_mal = len(df_cons[(df_cons.loc[:, 'VIVPARH_CV'] - df_cons.loc[:, '1cuart'] - df_cons.loc[:, '2cuart'] - df_cons.loc[:, '3mascuart']) != 0])
cuart_index = df_cons[(df_cons.loc[:, 'VIVPARH_CV'] - df_cons.loc[:, '1cuart'] - df_cons.loc[:, '2cuart'] - df_cons.loc[:, '3mascuart'])!=0].index
print('{} AGEBS tienen mal registrada la cantidad de cuartos.'.format(cuart_mal))
```

```python
def diff_num_cuart(x):
    
    '''
    x: es un dataframe con las columnas deseadas
    '''
    
    np.random.seed(123)
    
    x_vector = list(x)
    x_vector = [int(i) for i in x_vector]
    viviendas_tot = x_vector.pop(0)
    x_vector = np.array(x_vector)
    suma_var = sum(x_vector)
    
    if suma_var==viviendas_tot:
        return pd.Series(x_vector)
    
    else:
        x_vector = x_vector / suma_var * viviendas_tot
        xint = np.array([math.modf(i)[1] for i in x_vector])
        r = x_vector - xint
        
        if sum(r)!=1:
            r = r/r.sum()
            
        suma_r = sum(r)
        topup = np.random.choice(range(len(x_vector)), round(suma_r), p = r)
        xint[topup] = xint[topup] + 1
            
        return pd.Series(xint)
```

```python
df_cons[['1cuart', '2cuart', '3mascuart']] = df_cons[['VIVPARH_CV','1cuart', '2cuart', '3mascuart']].apply(diff_num_cuart, axis = 1)
df_cons[['1cuart', '2cuart', '3mascuart']].describe()
```
